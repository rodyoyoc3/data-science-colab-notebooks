```python

# Desinstalar posibles paquetes problemáticos
#!pip uninstall -y numpy scipy gensim thinc tsfresh

# Instalar versiones compatibles
#!pip install numpy==1.24.4 scipy==1.10.1 gensim==4.3.0 nltk==3.8.1 spacy==3.7.2 --quiet
#!python -m spacy download es_core_news_sm

!pip install gensim
from gensim import corpora
from gensim.models import LdaModel
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
import matplotlib.pyplot as plt

# Documentos por temática
documentos = [
    # Tecnología
    "La inteligencia artificial está revolucionando el sector industrial mediante la automatización inteligente, la mejora de procesos y el análisis predictivo en tiempo real.",
    "El aprendizaje automático se emplea para procesar grandes volúmenes de datos, optimizando la eficiencia en áreas como la logística, la ciberseguridad y el marketing digital.",

    # Salud
    "Las vacunas han permitido controlar epidemias globales y prevenir enfermedades infecciosas, fortaleciendo los sistemas de salud pública a nivel mundial.",
    "Los avances en biotecnología han facilitado diagnósticos más precisos, tratamientos personalizados y terapias innovadoras para enfermedades crónicas.",

    # Educación
    "La educación virtual ha transformado el aprendizaje mediante plataformas interactivas, cursos en línea y herramientas de evaluación automatizada.",
    "Los recursos digitales y las aulas híbridas permiten personalizar la enseñanza, adaptándose al ritmo de cada estudiante y facilitando el acceso a contenidos globales.",

    # Medio ambiente
    "El cambio climático ha provocado alteraciones en los ecosistemas, aumentando la frecuencia de fenómenos extremos como huracanes, sequías e incendios forestales.",
    "Las políticas de desarrollo sostenible buscan mitigar el impacto ambiental mediante energías limpias, reciclaje y conservación de la biodiversidad.",

    # Cultura y literatura
    "La literatura hispanoamericana ha influido profundamente en la cultura global, destacando por su riqueza narrativa y exploración de temas como la identidad, el exilio y la resistencia.",
    "Autores como Borges, García Márquez y Vargas Llosa han utilizado estilos como el realismo mágico y la metaficción para retratar contextos históricos y sociales complejos."
]

# Preprocesamiento
stop_es = stopwords.words('spanish')
textos_limpios = []
for doc in documentos:
    palabras = [pal.lower() for pal in doc.split() if pal.lower() not in stop_es and pal.isalpha()]
    textos_limpios.append(palabras)

# Diccionario y corpus
diccionario = corpora.Dictionary(textos_limpios)
corpus = [diccionario.doc2bow(texto) for texto in textos_limpios]

# Modelo LDA con 10 tópicos
modelo_lda = LdaModel(corpus=corpus, num_topics=10, id2word=diccionario, passes=20, random_state=42)

# Mostrar los tópicos
temas = modelo_lda.print_topics(num_words=14)
for idx, tema in temas:
    print(f"\n🌀 Tópico {idx}: {tema}")

# Graficar la distribución de tópicos por documento
topic_counts = [max(modelo_lda[doc], key=lambda x: x[1])[0] for doc in corpus]

plt.figure(figsize=(8, 5))
plt.hist(topic_counts, bins=range(11), align='left', rwidth=0.7, color='skyblue', edgecolor='black')
plt.title("Distribución de tópicos dominantes por documento")
plt.xlabel("Tópico dominante")
plt.ylabel("Cantidad de documentos")
plt.xticks(range(10))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

```

```python
!pip install scikit-learn nltk --quiet

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Documentos agrupados por pares temáticos (más extensos)
documentos = [
    # Tecnología
    "La inteligencia artificial está revolucionando la forma en que las empresas toman decisiones, automatizan procesos y predicen comportamientos de usuarios a través del análisis de datos.",
    "El aprendizaje automático, una rama de la IA, permite desarrollar sistemas que se adaptan al comportamiento humano y mejoran sus predicciones conforme se alimentan con nuevos datos.",

    # Salud
    "Los avances en medicina personalizada han permitido el desarrollo de tratamientos basados en el ADN del paciente, mejorando la eficacia terapéutica y reduciendo efectos secundarios.",
    "Las vacunas han sido claves en la erradicación de enfermedades y en la protección de comunidades frente a pandemias, reforzando la importancia de la salud pública global.",

    # Educación
    "La educación virtual ha crecido exponencialmente, impulsada por plataformas digitales que permiten el aprendizaje a distancia, flexible y personalizado para millones de estudiantes.",
    "El uso de inteligencia artificial en la educación permite adaptar contenidos al ritmo del estudiante, detectar dificultades de aprendizaje y automatizar la evaluación.",

    # Medio ambiente
    "El cambio climático se manifiesta en fenómenos extremos como huracanes más intensos, olas de calor y pérdida de biodiversidad, lo que exige políticas urgentes de mitigación.",
    "Las energías renovables como la solar y eólica son esenciales para reducir emisiones y avanzar hacia un modelo de desarrollo sostenible frente a la crisis ambiental actual.",

    # Cultura y literatura
    "La literatura hispanoamericana ha sido un vehículo de expresión política, identidad cultural y memoria histórica, con autores que exploran el realismo mágico y la crítica social.",
    "Obras de escritores como Borges, Allende y García Márquez han influido globalmente, combinando fantasía, historia y reflexión filosófica en contextos latinoamericanos complejos."
]

# Preprocesamiento: vectorización con stopwords en español
stop_es = stopwords.words('spanish')
vectorizador = TfidfVectorizer(stop_words=stop_es, max_features=1000)
X_tfidf = vectorizador.fit_transform(documentos)

# Aplicar NMF con 5 temas
modelo_nmf = NMF(n_components=10, random_state=42)
modelo_nmf.fit(X_tfidf)

# Mostrar los temas con palabras clave
vocabulario = vectorizador.get_feature_names_out()
for idx, topic in enumerate(modelo_nmf.components_):
    print(f"\n🧩 Tema {idx + 1}:")
    print(" + ".join([vocabulario[i] for i in topic.argsort()[-10:]]))
```

```python
!pip install pyldavis --quiet

import pyLDAvis.gensim_models
import pyLDAvis

# Preparar los datos para pyLDAvis
vis_data = pyLDAvis.gensim_models.prepare(modelo_lda, corpus, diccionario)

# Visualizar los resultados
pyLDAvis.display(vis_data)
```
