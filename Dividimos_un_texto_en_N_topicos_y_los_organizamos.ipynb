```python

# Desinstalar posibles paquetes problemticos
#!pip uninstall -y numpy scipy gensim thinc tsfresh

# Instalar versiones compatibles
#!pip install numpy==1.24.4 scipy==1.10.1 gensim==4.3.0 nltk==3.8.1 spacy==3.7.2 --quiet
#!python -m spacy download es_core_news_sm

!pip install gensim
from gensim import corpora
from gensim.models import LdaModel
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
import matplotlib.pyplot as plt

# Documentos por temtica
documentos = [
    # Tecnologa
    "La inteligencia artificial est revolucionando el sector industrial mediante la automatizacin inteligente, la mejora de procesos y el anlisis predictivo en tiempo real.",
    "El aprendizaje automtico se emplea para procesar grandes volmenes de datos, optimizando la eficiencia en reas como la logstica, la ciberseguridad y el marketing digital.",

    # Salud
    "Las vacunas han permitido controlar epidemias globales y prevenir enfermedades infecciosas, fortaleciendo los sistemas de salud pblica a nivel mundial.",
    "Los avances en biotecnologa han facilitado diagnsticos ms precisos, tratamientos personalizados y terapias innovadoras para enfermedades crnicas.",

    # Educacin
    "La educacin virtual ha transformado el aprendizaje mediante plataformas interactivas, cursos en lnea y herramientas de evaluacin automatizada.",
    "Los recursos digitales y las aulas hbridas permiten personalizar la enseanza, adaptndose al ritmo de cada estudiante y facilitando el acceso a contenidos globales.",

    # Medio ambiente
    "El cambio climtico ha provocado alteraciones en los ecosistemas, aumentando la frecuencia de fenmenos extremos como huracanes, sequas e incendios forestales.",
    "Las polticas de desarrollo sostenible buscan mitigar el impacto ambiental mediante energas limpias, reciclaje y conservacin de la biodiversidad.",

    # Cultura y literatura
    "La literatura hispanoamericana ha influido profundamente en la cultura global, destacando por su riqueza narrativa y exploracin de temas como la identidad, el exilio y la resistencia.",
    "Autores como Borges, Garca Mrquez y Vargas Llosa han utilizado estilos como el realismo mgico y la metaficcin para retratar contextos histricos y sociales complejos."
]

# Preprocesamiento
stop_es = stopwords.words('spanish')
textos_limpios = []
for doc in documentos:
    palabras = [pal.lower() for pal in doc.split() if pal.lower() not in stop_es and pal.isalpha()]
    textos_limpios.append(palabras)

# Diccionario y corpus
diccionario = corpora.Dictionary(textos_limpios)
corpus = [diccionario.doc2bow(texto) for texto in textos_limpios]

# Modelo LDA con 10 tpicos
modelo_lda = LdaModel(corpus=corpus, num_topics=10, id2word=diccionario, passes=20, random_state=42)

# Mostrar los tpicos
temas = modelo_lda.print_topics(num_words=14)
for idx, tema in temas:
    print(f"\n Tpico {idx}: {tema}")

# Graficar la distribucin de tpicos por documento
topic_counts = [max(modelo_lda[doc], key=lambda x: x[1])[0] for doc in corpus]

plt.figure(figsize=(8, 5))
plt.hist(topic_counts, bins=range(11), align='left', rwidth=0.7, color='skyblue', edgecolor='black')
plt.title("Distribucin de tpicos dominantes por documento")
plt.xlabel("Tpico dominante")
plt.ylabel("Cantidad de documentos")
plt.xticks(range(10))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

```

```python
!pip install scikit-learn nltk --quiet

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Documentos agrupados por pares temticos (ms extensos)
documentos = [
    # Tecnologa
    "La inteligencia artificial est revolucionando la forma en que las empresas toman decisiones, automatizan procesos y predicen comportamientos de usuarios a travs del anlisis de datos.",
    "El aprendizaje automtico, una rama de la IA, permite desarrollar sistemas que se adaptan al comportamiento humano y mejoran sus predicciones conforme se alimentan con nuevos datos.",

    # Salud
    "Los avances en medicina personalizada han permitido el desarrollo de tratamientos basados en el ADN del paciente, mejorando la eficacia teraputica y reduciendo efectos secundarios.",
    "Las vacunas han sido claves en la erradicacin de enfermedades y en la proteccin de comunidades frente a pandemias, reforzando la importancia de la salud pblica global.",

    # Educacin
    "La educacin virtual ha crecido exponencialmente, impulsada por plataformas digitales que permiten el aprendizaje a distancia, flexible y personalizado para millones de estudiantes.",
    "El uso de inteligencia artificial en la educacin permite adaptar contenidos al ritmo del estudiante, detectar dificultades de aprendizaje y automatizar la evaluacin.",

    # Medio ambiente
    "El cambio climtico se manifiesta en fenmenos extremos como huracanes ms intensos, olas de calor y prdida de biodiversidad, lo que exige polticas urgentes de mitigacin.",
    "Las energas renovables como la solar y elica son esenciales para reducir emisiones y avanzar hacia un modelo de desarrollo sostenible frente a la crisis ambiental actual.",

    # Cultura y literatura
    "La literatura hispanoamericana ha sido un vehculo de expresin poltica, identidad cultural y memoria histrica, con autores que exploran el realismo mgico y la crtica social.",
    "Obras de escritores como Borges, Allende y Garca Mrquez han influido globalmente, combinando fantasa, historia y reflexin filosfica en contextos latinoamericanos complejos."
]

# Preprocesamiento: vectorizacin con stopwords en espaol
stop_es = stopwords.words('spanish')
vectorizador = TfidfVectorizer(stop_words=stop_es, max_features=1000)
X_tfidf = vectorizador.fit_transform(documentos)

# Aplicar NMF con 5 temas
modelo_nmf = NMF(n_components=10, random_state=42)
modelo_nmf.fit(X_tfidf)

# Mostrar los temas con palabras clave
vocabulario = vectorizador.get_feature_names_out()
for idx, topic in enumerate(modelo_nmf.components_):
    print(f"\n Tema {idx + 1}:")
    print(" + ".join([vocabulario[i] for i in topic.argsort()[-10:]]))
```
